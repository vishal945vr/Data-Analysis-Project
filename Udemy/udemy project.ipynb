```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

#create data base

```python
#create dataset udemy
df=pd.read_csv("udemy.csv")
```

```python
df.head(5)
```

Data Clearning

```python
df.shape
```

```python
df.drop_duplicates("course_title")
```

```python
df.head(5)
```

```python
df.replace(True,1,inplace=True)
df.replace(False,0,inplace=True)

```

```python
df.isnull().sum()
```

```python
#data statictes
df.describe()
```

```python
df.drop_duplicates(inplace=True)
```

#Graphs

```python
plt.figure(figsize=(10,5))
sns.histplot(x="subject",data=df,hue="level",color="g")
plt.title("subject_level")
plt.ylabel("Value")
plt.xlabel("Subject")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(x="level",data=df,color="g",kde=True,palette=["blue","green"],hue="is_paid")
plt.show()
```

```python
sns.countplot(x="is_paid",data=df,order=[1,0],color="grey")
plt.show()
```

```python
# Customizing the pair plot
sns.pairplot(df, diag_kind="kde", height=2.5)
plt.show()


```

```python
plt.figure(figsize=(15,5))
sns.barplot(y="price",x="subject",data=df,color="orange")
plt.show()
```

```python
plt.figure(figsize=(20,5))
sns.lineplot(x="num_reviews",y="num_subscribers",data=df,color="green")
```
